```python

# Concatenate the DataFrames along the rows axis
merged_df = pd.concat([df1, df2, df3, df4, df5, df6, df7, df8, df9, df10, df11, df12])

# Reset the index of the merged DataFrame
merged_df.reset_index(drop=True, inplace=True)
merged_df.head()
```

```python
merged_df=merged_df.rename(columns={
    "ride_id":"Ride ID",
    "rideable_type":"Vehicle Type",
    "started_at":"Start Info",
    "ended_at":"End Info",
    "start_station_name":"Start Station Name",
    "start_station_id":"Start Station ID",
    "end_station_name":"End Station Name",
    "end_station_id":"End Station ID",
    "start_lat":"Start Latitude",
    "start_lng":"Start Longitude",
    "end_lat":"End Latitude",
    "end_lng":"End Longitude",
    "member_casual":"Membership Type"
})
merged_df.head()
```

```python
# Data Cleaning Process

# Checking for missing values in each column
missing_values = merged_df.isnull().sum()
missing_values
```

```python
# Checking for duplicate entries
duplicate_rows = merged_df.duplicated().sum()
duplicate_rows
```

```python
# Analyzing the data types of each column to see if any conversions are needed
data_types = merged_df.dtypes
data_types
```

```python
# Replace missing values
merged_df.fillna("Unknown", inplace=True)

# Convert date columns to datetime
merged_df['Start Info'] = pd.to_datetime(merged_df['Start Info'])
merged_df['End Info'] = pd.to_datetime(merged_df['End Info'])
```

```python
merged_df.dtypes
```

```python
# convert End Lat/Lng to float
# merged_df['End Latitude']= merged_df['End Latitude'].astype(float)
# merged_df['End Longitude']= merged_df['End Longitude'].astype(float)
merged_df['End Latitude'] = pd.to_numeric(merged_df['End Latitude'], errors='coerce')
merged_df['End Longitude'] = pd.to_numeric(merged_df['End Longitude'], errors='coerce')

merged_df.dtypes
```

```python
# save to csv
merged_df.to_csv('citibike2023.csv', index=False)
```
